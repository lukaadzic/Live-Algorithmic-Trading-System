## 1. Coin Flips: Expected Number of Tosses to get Two Heads in a Row

**Problem:** What is the expected number of fair coin tosses required to get two heads in a row (HH)?

**Analytical Solution:**
Let $E$ be the expected number of tosses to see HH.
- State 0: We start here. We toss a coin (takes 1 toss).
  - If Tails (prob 1/2), we are back to State 0. Tosses expected from here: $1 + E$.
  - If Heads (prob 1/2), we move to State 1.
- State 1: We have 1 Head. We toss again (takes 1 toss). 
  - If Tails (prob 1/2), we lose our streak. Tosses expected: $1 + E$.
  - If Heads (prob 1/2), we finish. Tosses expected: $1$.

Formula:
$E = \frac{1}{2}(1 + E) + \frac{1}{2}\left(\frac{1}{2}(2 + E) + \frac{1}{2}(2)\right)$

Solving for $E$:
$E = \frac{1}{2} + \frac{1}{2}E + \frac{1}{4}(2+E) + \frac{1}{4}(2)$

$E = \frac{1}{2} + \frac{1}{2}E + \frac{1}{2} + \frac{1}{4}E + \frac{1}{2}$

$E - \frac{3}{4}E = 1.5$

$\frac{1}{4}E = 1.5 \implies E = 6$

Let's simulate this in Python to verify.

```python
import random
import numpy as np

def simulate_HH():
    tosses = 0
    heads_in_a_row = 0
    while heads_in_a_row < 2:
        tosses += 1
        if random.random() < 0.5: # Heads
            heads_in_a_row += 1
        else: # Tails
            heads_in_a_row = 0
    return tosses

simulations = [simulate_HH() for _ in range(10000)]
print(f"Simulated Expected Tosses: {np.mean(simulations):.2f}")
print("Theoretical Expected Tosses: 6.00")
```

## 2. Dice Game: Stopping Rule

**Problem:** You roll a fair 6-sided die. You can choose to keep the score you roll, or re-roll. You can re-roll at most 2 times (so max 3 rolls total). What is your optimal strategy and expected value?

**Analytical Solution:**
- If you only have 1 roll left (Roll 3), your expected value is $\frac{1+2+3+4+5+6}{6} = 3.5$.
- With 2 rolls left (Roll 2), you should only keep a roll if it's greater than the expected value of the next roll (3.5). So you keep 4, 5, 6, and re-roll 1, 2, 3.
  - Expected value of Roll 2 = $\frac{1}{2}(4+5+6)/3 + \frac{1}{2}(3.5) = 2.5 + 1.75 = 4.25$.
- With 3 rolls left (Roll 1), you should keep a roll if it's greater than 4.25. So you keep 5 or 6, and re-roll 1, 2, 3, 4.
  - Expected value of Roll 1 = $\frac{2}{6}(5.5) + \frac{4}{6}(4.25) = 1.833 + 2.833 = 4.66$.

Let's simulate.

```python
def simulate_dice_game():
    """Simulates the dice game with optimal stopping rule."""
    roll1 = random.randint(1, 6)
    if roll1 >= 5:
        return roll1
        
    roll2 = random.randint(1, 6)
    if roll2 >= 4:
        return roll2
        
    roll3 = random.randint(1, 6)
    return roll3

results = [simulate_dice_game() for _ in range(100000)]
print(f"Simulated Expected Payoff: {np.mean(results):.3f}")
print("Theoretical Expected Payoff: ~4.667")
```

## 3. Card Draw: Expected Value in a Deck

**Problem:** Consider a standard 52-card deck. I deal you cards one by one. You can say 'stop' at any time. Your payout is the number of red cards dealt minus the number of black cards dealt. What is your optimal strategy?

**Analytical Solution:**
It's a martingale process. The expected return of this game is exactly 0 regardless of the stopping strategy. By the Optional Stopping Theorem, since the game is bounded and must stop eventually, the expected value of stopping is equal to the initial expected value, which is 0.
