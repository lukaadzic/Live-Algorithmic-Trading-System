# Paper Trading Strategy & Risk Analysis

This notebook analyzes the `paper_trading_history.csv` generated by the live multi-asset simulation. We look at the absolute equity growth, portfolio drawdowns, and per-asset unrealized PnL to evaluate strategy quality and risk.

```python
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

# Load paper trading history
df = pd.read_csv('../paper_trading_history.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])

# 1. Overall Equity Curve
plt.figure(figsize=(10, 5))
plt.plot(df['timestamp'], df['total_equity'], color='b', label='Total Equity')
plt.title('Live Simulation Equity Curve')
plt.xlabel('Time')
plt.ylabel('Equity (USD)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 2. Maximum Drawdown (Risk Management Indicator)
df['peak'] = df['total_equity'].cummax()
df['drawdown'] = (df['total_equity'] - df['peak']) / df['peak']

plt.figure(figsize=(10, 5))
plt.plot(df['timestamp'], df['drawdown'], color='red', label='Drawdown')
plt.fill_between(df['timestamp'], df['drawdown'], 0, color='red', alpha=0.3)
plt.title('Portfolio Drawdown')
plt.xlabel('Time')
plt.ylabel('Drawdown (%)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 3. Per-Asset Unrealized PnL
assets = ['AAPL_pnl', 'MSFT_pnl', 'AMZN_pnl', 'NVDA_pnl']
available_assets = [col for col in assets if col in df.columns]

if available_assets:
    plt.figure(figsize=(10, 5))
    for asset in available_assets:
        plt.plot(df['timestamp'], df[asset], label=asset)

    plt.title('Per Asset Unrealized PnL Overlay')
    plt.xlabel('Time')
    plt.ylabel('PnL (USD)')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)
    plt.show()
```
